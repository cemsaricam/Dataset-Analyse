```python
plt.figure(figsize=(12,6))

plt.plot(vs.Tarih,vs.enflasyon,color="red") 

plt.plot(vs.Tarih,vs.issizlik,color="blue")

plt.xlabel("Tarih")

plt.ylabel("Enflasyon ve İşsizlik Oranı")

plt.title("Türkiye'nin Enflasyon ve İşsizlik Oranları")

plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)   
plt.plot(vs.Tarih,vs.enflasyon,color="r") 
plt.xlabel("Tarih")
plt.ylabel("Enflasyon Oranı")
plt.title("Türkiye'nin Enflasyon Oranları")

plt.subplot(2,2,2)
plt.plot(vs.Tarih,vs.issizlik,color="blue")
plt.xlabel("Tarih")
plt.ylabel("İşsizlik Oranı")
plt.title("Türkiye'nin İşsizlik Oranları")

plt.show()
```

```python
f=plt.figure(figsize=(10,4))

axes=f.add_axes([0.1,0.1,0.9,0.9])

axes.plot(vs.Tarih,vs.enflasyon)

axes.set_xlabel("Tarih")

axes.set_ylabel("Enflasyon")

axes.set_title("Türkiye'nin Enflasyon Oranları")

plt.show()
```

```python
f=plt.figure(figsize=(12,4)) 


#koordinatlar
axes1=f.add_axes([0.1,0.1,0.9,0.9]) 

axes2=f.add_axes([0.6,0.3,0.3,0.3])


#verilerin grafikte nasıl yer alacağı.
axes1.plot(vs.Tarih,vs.enflasyon,color="blue")

axes2.plot(vs.Tarih,vs.issizlik,color="red")


#x ekseni başlıkları
axes1.set_xlabel("Tarih")
axes2.set_xlabel("Tarih")

#y ekseni başlıkları
axes1.set_ylabel("Enflasyon")
axes2.set_ylabel("işsizlik")

#Grafik başlıkları
axes1.set_title("Türkiye'nin Enflasyon Oranları")
axes2.set_title("Türkiye'nin İşsizlik Oranları")

plt.show()
```
